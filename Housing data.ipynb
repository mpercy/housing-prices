```python
def ToCountyDataFrame(df, county_name):
    s = df.loc[df['RegionName'] == county_name] \
             .set_index("RegionName") \
             .drop(sc.columns[[0, 1, 2, 3, 4, 5, 6]], axis=1) \
             .rename(columns = lambda x: pd.to_datetime(x)) \
             .transpose()
    s.columns = ['Price']
    return s
```
